```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
from colorama import Fore
```

```python
data=pd.read_excel('/Users/omarhassan/Documents/GitHub/BigMart-Analysis/Dataset/final_dataset.xlsx',sheet_name='Original_Imputed')
data.head(10)
```

```python
data.isnull().sum()
```

### Exploratory Data Analysis

```python
fig,ax=plt.subplots(figsize=(20,10))
ax.barh(data['Item_Type'],data['Item_Outlet_Sales'])
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
ax.grid(color ='black',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
ax.invert_yaxis()
plt.show()
```

It is seen that Household type Items are the most sold.  
Fruits and vegetables are the next most sold.  
Followed by Snack Foods.  
  
Items labeled Others are the least sold.  
Then comes Seafood, being the second least sold.  
Followed by Hard Drinks and then comes Baking Goods.

```python
plt.bar(data['Item_Fat_Content'],data['Item_Outlet_Sales'])
```

Low Fat products are sold more than Regular products.
